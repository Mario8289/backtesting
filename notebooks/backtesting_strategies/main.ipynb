```python
df = result.df
```

```python
df.groupby('event_type').agg({'trading_session': 'count'})
```

```python
md = df[df['event_type'] == 'market_data'].sort_index()
t = df[df['event_type'] == 'trade'].sort_index()
```

```python
from bokeh.plotting import figure, show, output_notebook

output_notebook()

# create a new plot with a title and axis labels
p = figure(
    title="Bitcoin DCA", 
    x_axis_type='datetime',
    x_axis_label='x', 
    y_axis_label='y',
    width=1000,   # Set the width in pixels
    height=400   # Set the height in pixels
)

# add a line renderer with legend and line thickness to the plot
p.line(md.index, md['unrealised_pnl_cum'], legend_label="Market Data", line_width=2)
p.scatter(t.index, t['unrealised_pnl_cum'], legend_label="Trades", line_width=2, size=8, color='green')

# show the results
show(p)
```
